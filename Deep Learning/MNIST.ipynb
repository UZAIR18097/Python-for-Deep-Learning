```python
import tensorflow as tf
print(tf.__version__)
```

# load model

```python
mnist = tf.keras.datasets.fashion_mnist
(training_images, training_labels), (test_images, test_labels) = mnist.load_data()
training_images  = training_images / 255.0
test_images = test_images / 255.0
```

```python
print(training_images.ndim)
print(training_images.shape)
print(training_images.dtype)
```

# prepare model

```python
model = tf.keras.models.Sequential([tf.keras.layers.Flatten(),               
tf.keras.layers.Dense(512, activation=tf.nn.relu),                           
tf.keras.layers.Dense(10, activation=tf.nn.softmax)])
```

# Assign optimizer and loss function

```python
model.compile(optimizer = tf.keras.optimizers.Adam(), loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

# training

```python
model.fit(training_images, training_labels, epochs=10, batch_size=128)
```

# evaluate

```python
model.evaluate(test_images, test_labels)
```

# Mnist Digit

```python
from tensorflow.keras.datasets import mnist
```

```python
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
train_images.shape
```

```python
len(train_images)
```

```python
train_labels[0]
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers
```

```python
network = models.Sequential()
network.add(layers.Dense(512, activation='relu', input_shape=(28 * 28,)))
network.add(layers.Dense(10, activation='softmax'))#sigmoid for binary classification,softmax for more than two class
```

```python
network.compile(optimizer='rmsprop',
loss='categorical_crossentropy',
metrics=['accuracy'])
```

```python
train_images = train_images.reshape((60000, 28 * 28))
train_images = train_images.astype('float32') / 255
test_images = test_images.reshape((10000, 28 * 28))
test_images = test_images.astype('float32') / 255
```

```python
from tensorflow.keras.utils import to_categorical
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

```python
train_labels
```

```python
test_labels
```

```python
network.fit(train_images,train_labels,epochs=5,batch_size=128)
```

```python
test_loss, test_acc = network.evaluate(test_images, test_labels)
print('test_acc:', test_acc)
```
